Problem Statement:
The study aims to investigate and analyze the potential risk factors associated with shark attacks in coastal regions. The research focuses on three distinct hypotheses to better understand the circumstances surrounding shark-human interactions.

Hypothesis #1:
Individuals participating in surfing activities are more likely to experience shark attacks of a fatal nature compared to those engaged in other water-related activities, indicating a potential correlation between surfing and an increased probability of encountering sharks.

Hyypothesis #2:
#The likelihood of shark attacks is higher during the early morning hours, specifically between 6:00 AM and 11:00 AM.

Hypothesis #3:
There is a greater probability of a shark attack occurring when the shark is provoked as opposed to unprovoked, 
specifically during the water activity being practiced at the time of the attack.


For Hypothesis #1 we need the columns: "killed" and "activity".

For Hypothesis #2 we need the column: "time".

For Hypothesis #3 we need the column: "type".




```python
shark_attack_df=shark_attack_df.drop(shark_attack_df.index[6920:])
shark_attack_df = shark_attack_df.drop(['href formula','href', 'Case Number', 'Case Number.1',
       'original order', 'Unnamed: 21', 'Unnamed: 22'],axis=1)
```

```python
shark_attack_df.columns = shark_attack_df.columns.str.lower()
shark_attack_df
```

```python
shark_attack_df = shark_attack_df.rename({'unnamed: 11':'killed'},axis =1)
shark_attack_df
```
